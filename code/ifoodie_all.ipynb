# 爬取所有頁面資料

```python
driver = webdriver.Chrome("chromedriver")
#driver = webdriver.Chrome(ChromeDriverManager().install())
driver.set_window_size(600,800)

ifoodie3 = pd.DataFrame()
iter_id = 80

for link in tqdm(links[80:]):

    URL = link
    driver.get(URL)
    driver.implicitly_wait(10)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
    time.sleep(5)

    soup = BeautifulSoup(driver.page_source,"html.parser")
    
    # 店名、連結
    names = soup.find_all("a", class_ = "jsx-1156793088 title-text")
    name_list = []
    link_list = []
    for name in names:
        name_list.append(name.text)
        link_list.append("https://ifoodie.tw"+name["href"])

    # 地址
    addresses = soup.find_all("div", class_ = "jsx-1156793088 address-row")
    address_list = []
    for address in addresses:
        address_list.append(address.text)

    # 評分
    scores = soup.find_all("div", class_ = "jsx-1156793088 review-row")
    score_list=[]
    for tempscore in scores:
        score = tempscore.find("div", class_ = "jsx-2373119553 text")
        if score != None:
            score_list.append(float(score.text))
        else:
            score_list.append(np.nan)

    # 評論數
    reviews = soup.find_all("a", class_ = "jsx-1156793088 review-count")
    review_list = []
    for review in reviews:
        #review_list.append(review.text)
        review_list += re.findall(r'\d+', review.text)
    review_list = list(map(int,review_list))

    # 均消
    prices = soup.find_all("div", class_ = "jsx-1156793088 avg-price")
    price_list = []
    for price in prices:
        price_list += re.findall(r'\d+', price.text)
    price_list = list(map(int,price_list))

    # 餐廳種類
    categories = soup.find_all("a", class_ = "jsx-1156793088 category")
    category_list=[]
    category_set = str()
    for category in categories:
        if category.text == "附近餐廳":
            category_list.append(category_set.rstrip(",")) #刪掉最後一個逗號
            category_set = str()
            next
        else:
            category_set = category_set + category.text + ","
    category_list.remove("") #第一個是空字串
    category_list.append(category_set.rstrip(",")) #最後一個沒有被加進list

    # index（用做圖檔名稱）
    indeies = soup.find_all("span", class_ = "jsx-1156793088 index")
    index_list = []
    for index in indeies:
        index_list.append(dist(link) + index.text + "png")

    # 下載封面圖檔
    covers = soup.find_all("img", {'class':['jsx-1156793088 cover','jsx-1156793088 cover lazyload','jsx-1156793088 cover lazyloaded','jsx-1156793088 cover ls-is-cached lazyloaded']})
    cover_list = []
    for cover in covers:
        if ("data:image/gif;" or "data:image/svg") in str(cover):
            cover_list.append(cover["data-src"].rstrip("360")+"600") # 原圖檔s360，想下載較大尺寸，故改成s600
        else:
            cover_list.append(cover["src"].rstrip("360")+"600")

    for i in range(len(cover_list)):
        if "data:image/svg" in cover_list[i]:
            cover_list[i] = ""
        elif "data:image/gif" in cover_list[i]:
            cover_list[i] = ""
        else:
            img = requests.get(cover_list[i]).content
            with open("image/"+index_list[i], 'wb') as handler:
                handler.write(img)

    # dataframe
    one_page_data=pd.DataFrame(list(zip(name_list,address_list,score_list,review_list,price_list,category_list,index_list,cover_list,link_list)),
                            columns=["name","address","score","num_score","ave_price","category","image_file","image_source","link"])
    ifoodie3 = pd.concat([ifoodie3, one_page_data],ignore_index=True)
    ifoodie3.to_csv(("ifoodie3.csv"), index=False, encoding="utf-8")

    iter_id += 1

    time.sleep(60)
    driver.execute_script("window.scrollTo(0, 100)")
    time.sleep(60)

driver.quit()
```

```python
# 爬取所有營業時間

driver = webdriver.Chrome(ChromeDriverManager().install())
#driver = webdriver.Chrome("chromedriver")

for i in tqdm(range(len(links))):
    
    URL = links[i]
    driver.get(URL)
    driver.implicitly_wait(10)
    driver.find_element(By.CLASS_NAME,"jss41.icon-expand").click()

    soup = BeautifulSoup(driver.page_source,"html.parser")
    hours = soup.find_all("div", class_ = "jss76 jss116 jss119 jss124 jss125")
    for hour in hours:
        day_hour = hour.text
        if "一" in day_hour:
            ifoodie_all.iloc[i,9] = re.sub("星期一 ", "", day_hour)
        elif "二" in day_hour:
            ifoodie_all.iloc[i,10] = re.sub("星期二 ", "", day_hour)
        elif "三" in day_hour:
            ifoodie_all.iloc[i,11] = re.sub("星期三 ", "", day_hour)
        elif "四" in day_hour:
            ifoodie_all.iloc[i,12] = re.sub("星期四 ", "", day_hour)
        elif "五" in day_hour:
            ifoodie_all.iloc[i,13] = re.sub("星期五 ", "", day_hour)
        elif "六" in day_hour:
            ifoodie_all.iloc[i,14] = re.sub("星期六 ", "", day_hour)
        elif "日" in day_hour:
            ifoodie_all.iloc[i,15] = re.sub("星期日 ", "", day_hour)

    ifoodie_all.to_csv(("ifoodie_operating.csv"), index=False, encoding="utf-8")
    
    if i%15==0:
        time.sleep(60)

driver.quit()
```
